```python
import pandas as pd
import glob
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print("Current working directory:", os.getcwd())
# Define the path pattern to locate .psv files in the folder
file_pattern = "CC_2019_Sepsis/training_setA/*.psv"
files = glob.glob(file_pattern)
print("Files found:", files)
# Option 1: Read all files into a list of DataFrames
dfs = [pd.read_csv(file, sep='|') for file in files]

# Combine all DataFrames into a single one (if desired)
combined_df = pd.concat(dfs, ignore_index=True)

print(combined_df)
```

```python
sns.scatterplot(data=combined_df, x='Hgb', y='Platelets', hue='SepsisLabel')
plt.title("Scatter Plot of Variable1 vs Variable2 by Sepsis Status")
plt.xlabel("Hgb")
plt.ylabel("Platelets")
plt.show()
```

```python
for col in combined_df.columns:
    if pd.api.types.is_numeric_dtype(combined_df[col]):
        combined_df[col] = combined_df[col].fillna(combined_df[col].median())

display(combined_df)
```

```python
#used minmax scaler check with Collin
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
combined_df[combined_df.select_dtypes(include=['float64', 'int64']).columns] = scaler.fit_transform(combined_df.select_dtypes(include=['float64', 'int64']))

display(combined_df)
```

```python
from sklearn.model_selection import train_test_split
X = combined_df.drop('SepsisLabel', axis=1)
y = combined_df['SepsisLabel']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=66)

print("Training Features:\n", X_train)
print("Testing Features:\n", X_test)
print("Training Labels:\n", y_train)
print("Testing Labels:\n", y_test)
```

```python
from collections.abc import Sequence
import numpy as np
def to_xy(df, target):
    result = []
    for x in df.columns:
        if x != target:
            result.append(x)
    # find out the type of the target column. 
    target_type = df[target].dtypes
    target_type = target_type[0] if isinstance(target_type, Sequence) else target_type
    # Encode to int for classification, float otherwise. TensorFlow likes 32 bits.
    if target_type in (np.int64, np.int32):
        # Classification
        dummies = pd.get_dummies(df[target])
        return df[result].values.astype(np.float32), dummies.values.astype(np.float32)
    else:
        # Regression
        return df[result].values.astype(np.float32), df[target].values.astype(np.float32)
```

```python
import tensorflow as tf
import keras
from keras import layers
import io
import requests
from sklearn import metrics
model = keras.Sequential(
    [
        layers.Dense(50, input_dim=X_train.shape[1], activation="relu", name="layer1"),
        layers.Dense(25, activation="relu", name="layer2"),
        layers.Dense(y_train.shape[1], activation='softmax')
    ]
)

model.compile(loss='categorical_crossentropy', optimizer='adam')

model.fit(X_train,y_train,verbose=2,epochs=100)
```
